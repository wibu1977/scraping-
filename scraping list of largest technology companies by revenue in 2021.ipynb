```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable plainrowheads')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
rows =table.find_all('tr')
```

```python
for row in rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [row_data[0].text.strip()] + [data.text.strip() for data in row_data[2:7]]
    print(individual_row_data)
    length = len(df)
    df.loc[length]= individual_row_data 
```
